**Шифр Вермана:**

В 1917 году телеграфист Гильберт Вернам изобрёл шифр (его также называют одноразовым блокнотом), который основан на побитовом исключающем ИЛИ. Если коротко и просто, то на каждую букву вашего сообщения накладывается другая маскирующая буква, которая делает исходную букву нечитаемой. Без знания ключа, расшифровать его невозможно (доказано Клодом Шенноном).

**Принцип работы:**

Как уже говорилось выше, щифр Вермана работает по принципу операции XOR(исключающие или). Если все сигналы равны между собой (0-0 или 1-1 или 0-0-0 и т.д.), то на выходе получаем 0. Если сигналы не равны (0-1 или 1-0 или 1-0-0 и т.д.), то на выходе получаем 1. Кодируется не сами символы, а их числовое обозначение в той или иной системе.

**Минусы шифра Вернама:**


 Если бы у этого шифра не было минусов, все бы пользовались только им, но сейчас этот шифр используется редко. Всё дело в том, что им очень неудобно пользоваться, чтобы выполнить все условия.


Чтобы этот шифр нельзя было взломать, нужно, чтобы секретный ключ был у обеих сторон и при этом его нельзя было перехватить. Но если так можно передать секретный ключ, то вместо него можно передать и само сообщение.


Даже если удастся заранее передать набор ключей каждой стороне для будущих шифровок, их нужно держать в полной физической безопасности, чтобы нельзя было их украсть, подсмотреть, потерять или перепутать.


1.   В качестве ключа должна быть абсолютно случайная последовательность — вероятно, придется стоять и подбрасывать кубик, чтобы сгенерировать такую;
2.   Для передачи необходим защищенный канал — сомнительно, что он всегда имелся под рукой во времена Первой мировой войны;
3.   Если третья сторона сможет каким-то образом узнать послание, она легко и восстановит ключи, и подменит сообщение;
4.   Требуется надежное уничтожение страницы блокнота — сжечь ее и съесть пепел, тогда враг точно не узнает, что было зашифровано. Именно из-за таких требований к безопасности ключа этот шифр сейчас используется очень редко.

**1. Шифрование**

Тест №1

```python
text= input("Введите кодируемое сообщение ").upper()
key=input("Введите кодовый текст ").upper()
len_key=len(key)#длина ключа
ord_cipher=[ord(text[i])^ord(key[i%len_key]) for i in range(len(text))]#шифр в 10ой системе
bit_cipher=[bin(i) for i in ord_cipher]#шифр в 2ой системе
if len(key)!=len(text):
  print("Ваш шифр может быть подвержен взлому , длина соотбщения для успеха "\
        "должна быть равна тексту , если хотите это исправить "\
        "перезагрузите программу")
print("Шивровка в виде чисел : ",*ord_cipher)
print("Шивровка в виде XOR битов :  ",*bit_cipher)
print("Зашифрованое сообщение : ",end="")
for i in ord_cipher:
    print(chr(i),end="")
```

**2. Дешифрование**

Работает тот же алгоритм:

Если у нас есть закодированное сообщение и ключ от него, то раскодировать его не составит труда. Мы просто пишем алгоритм, который применяет операцию XOR как бы в обратном порядке к каждому биту сообщения. Например:

Если в ключе шифрования в каком-то бите стоит 1, то мы берём этот же бит в зашифрованном сообщении и меняем его. Если был 1, то меняем на 0. Если был 0, то меняем на 1. Если в ключе шифрования в каком-то бите стоит 0, то этот же бит в зашифрованном сообщении мы не меняем. Проходим так по каждому биту сообщения и получаем исходный текст. Компьютер может совершать миллиарды таких операций в секунду. Главное — иметь на руках ключ для расшифровки.

Тест №1

```python
text= input("Введите кодируемое сообщение ").split()

key=input("Введите кодовый текст ").upper()
len_key=len(key)#длина ключа
ord_cipher=[int(text[i])^ord(key[i%len_key]) for i in range(len(text))]#шифр в 10ой системе
bit_cipher=[bin(i) for i in ord_cipher]#шифр в 2ой системе
print("Дешифвровка в виде чисел : ",*ord_cipher)
print("Дешифвровка в виде XOR битов :  ",*bit_cipher)
print("Дешифрованое сообщение : ",end="")
for i in ord_cipher:
    print(chr(i),end="")
```

**3. Взлом шифра Вермана**

Почему этот шифр невзламываемый Для начала договоримся, что под взломом мы понимаем прочтение этого сообщения без ключа. Если бы у нас был ключ, мы бы прочитали это сообщение почти сразу, и это уже не взлом.

Теперь посмотрим, почему без ключа этот шифр невозможно взломать.

1.   Каждый бит нашего исходного сообщения шифруется соответствующим битом, который берётся из ключа шифрования.
2.   Ключ шифрования — это случайные биты, такой «цифровой шум». Он не имеет смысла и в нём нет никакой логики. Каждый следующий бит может быть каким угодно.
3.   Шифрование происходит на самом низком уровне — на уровне битов. Мы даже не знаем, что перед нами: буквы, цифры, числа, картинки или аудио. Просто какой-то набор битов, которые выглядят как цифровой шум. Единственный способ расшифровать целое сообщение — это получить целый ключ. Если мы получим лишь часть ключа, мы не сможем угадать или восстановить недостающую часть. Сколько ключа у нас есть — столько битов сообщения мы и расшифруем. Нет ключа — нет расшифровки. Но есть некоторые уязвимости при неправильном использовании шифра Вермана: если длина ключа меньше длины сообщения, то это означает, что ключ будет использован несколько раз, что нарушает основной принцип шифра Вернама. Использование одного и того же ключа дважды может привести к разгадыванию шифра с помощью анализа частотности и других методов криптоанализа.


**Описание программы:** 


Данная программа реализует взлом шифра Вермана с помощью метода Касиски. Программа запрашивает на вход зашифрованный текст (в виде чисел в системе ASCII) и длину ключа, затем она находит возможные длины ключа с помощью метода Касиски, выбирает наиболее вероятные длины ключа, находит сам ключ и дешифрует сообщение.


**Описание функций:**


1.   Функция decrypt_vernam принимает на вход зашифрованный текст и ключ, и возвращает расшифрованный текст, используя операцию XOR.
2.   Функция kasiski_analysis проводит анализ частоты повторения триграмм в зашифрованном тексте и находит расстояния между повторяющимися триграммами. Затем она находит общие делители для этих расстояний и считает количество вхождений каждого делителя.
3.   Функция find_key_length использует результаты анализа Касиски для нахождения наиболее вероятной длины ключа. Она находит общие делители между расстояниями между повторяющимися триграммами, затем считает количество вхождений каждого делителя и возвращает пять наиболее часто встречающихся длин ключа.
4.   Функция find_key использует найденную длину ключа для нахождения самого ключа. Она разбивает зашифрованный текст на блоки длиной в длину ключа, затем для каждого блока находит символ, который наиболее часто встречается в этом блоке, и предполагает, что этот символ соответствует символу пробела в расшифрованном тексте. Таким образом, она находит символы ключа.
5.   Затем программа запрашивает у пользователя ввод зашифрованного сообщения и длины ключа, находит ключ и дешифрует сообщение, выводя результат на экран.

```python
from collections import Counter,defaultdict
from math import gcd

def decrypt_vernam(ciphertext, key):
    plaintext = ''
    for i in range(len(ciphertext)):
        plaintext += chr(ciphertext[i] ^ key[i % len(key)])
    return plaintext

def kasiski_analysis(ciphertext):
    distances = []
    for i in range(len(ciphertext) - 3):
        substring = ciphertext[i:i+3]
        try:
            distance = ciphertext[i+3:].index(substring) + 3
            distances.append(distance)
        except ValueError:
            pass
    factors = []
    for distance in distances:
        for i in range(2, distance):
            if distance % i == 0:
                factors.append(i)
                factors.append(distance // i)
    factor_counts = Counter(factors)
    return factor_counts

def find_key_length(ciphertext,max_key_length=5):
    distances = []
    # находим расстояния между повторяющимися триграммами
    for i in range(len(ciphertext) - 2):
        trigram = ciphertext[i:i + 3]
        try:
            distance = ciphertext[i + 3:].index(trigram) + 3
            distances.append(distance)
        except ValueError:
            pass

    # находим общие делители расстояний
    factors = []
    for i in range(2, max_key_length + 1):
        for distance in distances:
            if distance % i == 0:
                factors.append(i)

    # находим наиболее часто встречающиеся длины ключа
    factor_counts = Counter(factors)
    key_lengths = factor_counts.most_common(5)

    return [key[0] for key in key_lengths]


def find_key(ciphertext, key_length):
    key = [0] * key_length
    for i in range(key_length):
        block = []
        for j in range(i, len(ciphertext), key_length):
            block.append(ciphertext[j])
        block_counts = Counter(block)
        top_char = block_counts.most_common(1)[0][0]
        key[i] = top_char ^ 32  # предположим, что пробел является наиболее распространенным символом
    return key

ciphertext = [int(i) for i in input().split()]
key_length = find_key_length(ciphertext)
if key_length==[]:
    print("Программе не удалось найти длинну ключа попробуйте предположить его максимальную длинну ")
    key_length_=int(input())+1
    key_length=[i for i in range(1,key_length_)]

for i in key_length:
    key = find_key(ciphertext, i)
    print('Key:', key)

    # decrypt ciphertext
    plaintext = decrypt_vernam(ciphertext, key)
    print('Plaintext:', plaintext)
```
